```python
for f in df2.columns:
    if df2[f].dtype=='object':
        #print(f)
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(df2[f].values))
        df2[f] = lbl.transform(list(df2[f].values))
       
```
