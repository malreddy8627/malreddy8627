```python
purchase = pd.read_csv("train.csv")
```

```python
purchase = purchase.drop("QuoteNumber", axis=1)
```

```python
purchase['Date'] = pd.to_datetime(pd.Series(purchase['Original_Quote_Date']))
```

```python
purchase.head()
```

```python
purchase['weekday'] = purchase['Date'].dt.dayofweek
```

```python
purchase['year'] = purchase['Date'].dt.year
```

```python
purchase['month'] = purchase['Date'].dt.month
```

```python
purchase['wkofyear'] = purchase['Date'].dt.weekofyear
```

```python
purchase = purchase.drop(['Date', 'Original_Quote_Date'], axis=1)
```

```python
purchase.columns
```

```python
purchase.dtypes
```

```python
for i in purchase.columns: 
   print (purchase[i].name , purchase[i].dtype)
```

```python
purchase = purchase.fillna(-99)
```

```python
for f in purchase.columns:
    if purchase[f].dtype=='object':
        #print(f)
        lbl = LabelEncoder()
        lbl.fit(list(purchase[f].values))
        purchase[f] = lbl.transform(list(purchase[f].values))
```

```python
#lbl = LabelEncoder()
#lbl.fit(list(purchase["Field12"].values))
```

```python
for i in purchase.columns: 
   print (purchase[i].name , purchase[i].dtype)
```

```python
purchase["Field10"].value_counts()
```

```python
target = purchase["QuoteConversion_Flag"]
```

```python
purchase = purchase.drop("QuoteConversion_Flag", axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(purchase, target, test_size=0.3, random_state = 12345)
```

```python
clf = xgb.XGBClassifier(n_estimators=1000,
                        nthread=-1,
                        max_depth=12,
                        learning_rate=0.01,
                        subsample=0.6,
                        colsample_bytree=0.5,
                       silent=False
                       )
```

```python
clf.fit( X_train, y_train)
```

```python
y_target = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_target)
```
